### Analisis Datos Tráfico Barcelona

Importamos librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 2018

### Enero

Cargamos datos del csv desde url

```python
url_fichero = 'https://opendata-ajuntament.barcelona.cat/resources/auto/transit/2018_01_Gener_ITINERARIS_ITINERARIS.csv'
dfGener_2018 = pd.read_csv(url_fichero)
```

Mostramos columnas y valores del dataframe creado

```python
print(dfGener_2018.columns)
dfGener_2018.head()
dfGener_2018
```

#### Examinamos los datos

Tipo

```python
print(dfGener_2018.dtypes)
```

Cantidad de nulos = 0

```python
dfGener_2018.isnull().sum()
```

Examinamos los valores de tempsActual

```python
dfGener_2018['tempsActual'].value_counts()
```

Eliminamos las filas con tempsActual 0. Ya que vamos a obtener la mediana de este campo

```python
dfGener_2018 = dfGener_2018.drop(dfGener_2018[dfGener_2018['tempsActual'] == 0].index)
dfGener_2018
```

Eliminamos las columnas que no son de interes

```python
dfGener_2018 = dfGener_2018.drop(columns = ['idTram','infoDisponible','tempsPrevist','tempsRecorregutFutur','factorReferenciaActual','tendencia'])
dfGener_2018
```

Renombramos columnas de data por fecha

```python
dfGener_2018 = dfGener_2018.rename(columns = {'data':'fecha'})
```

#### Formateamos el campo fecha

```python
dfGener_2018['fecha'] = pd.to_datetime(dfGener_2018['fecha'], format='%Y%m%d%H%M%S')
dfGener_2018['fecha']
```

#### Agrupamos

Añadimos campo fecha como indice

```python
dfGener_2018 = dfGener_2018.set_index('fecha')
dfGener_2018
```

Agrupamos los registros por día y obtenemos la suma del tiempo en segundos por día

```python
dfTA_Gener_2018 = dfGener_2018.resample('D').median()
dfTA_Gener_2018
```

Compobamos que el dataframe tiene todos los dias del mes que le corresponse

```python
len(dfTA_Gener_2018)
```

Gráfico de visualización

```python
plt.plot(dfTA_Gener_2018)
plt.show()
```
