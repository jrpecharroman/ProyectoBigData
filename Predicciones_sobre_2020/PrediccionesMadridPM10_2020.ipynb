# Predicción de valores PM10 en Madrid en base a datos de Meteorología y Tráfico durante el 2020

### En este Notebook vamos a usar los datos del 2020 en Madrid para aplicar sobre ellos las predicciones de nuestro modelo entrenado y guardado para PM10. Compararemos los resultados predichos con los valores reales que hubo de PM10.

```python
#Cargamos las librerías que vamos a necesitar en el proceso

import numpy as np
import pandas as pd
from numpy import set_printoptions
#Visualización
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns
#Ensamblamiento Regresión
from sklearn.ensemble import ExtraTreesRegressor
#Metricas
from sklearn.metrics import mean_squared_error
#Serializar (guardar, cerrar y abrir modelo)
from joblib import load
```

## Carga y unión de datasets del 2020

```python
#Cargamos el dataset de datos de Calidad Madrid 2020

url_fichero_cal = '../Contaminacion/datosContaminacionMadrid/JupiterNoteBook/contaminacionMadrid2020.csv'
dfContaminacionMadrid = pd.read_csv(url_fichero_cal, delimiter=';', na_values='?')
dfContaminacionMadrid = dfContaminacionMadrid.set_index('date')
```

```python
#ahora cargamos el dataset de datos de Meteorología en Madrid 2020

url_fichero_met = '../Meteorologia/df_sinhoras_estacion_datos_diarios_mad_2020.csv'
dfMeteoMadrid = pd.read_csv(url_fichero_met)
dfMeteoMadrid = dfMeteoMadrid.set_index('fecha')
dfMeteoMadrid.index.rename('date', inplace=True)
```

```python
#por último cargamos el dataset con los datos de Trafico Madrid 2020

url_fichero_traf = '../Tráfico/Madrid/trafico_mad_2020.csv'
dfTraficoMadrid = pd.read_csv(url_fichero_traf, delimiter=',', na_values='?')
dfTraficoMadrid = dfTraficoMadrid.set_index('fecha')
dfTraficoMadrid.index.rename('date', inplace=True)
```

```python
#Ahora unimos los 3 datasets en uno solo

dfMergeado = pd.merge(dfContaminacionMadrid, dfMeteoMadrid, left_index=True, right_index=True)
dfMergeado = pd.merge(dfMergeado, dfTraficoMadrid, left_index=True, right_index=True)
print("columnas primer df -->",dfMergeado.columns)


# le eliminamos las columnas innecesarias o duplicadas
dfMergeado_2020 = dfMergeado.drop(columns=['o3','pm25','so2','indicativo','season','ocupacion','season_name','nombre','year','carga','month','week','provincia'])
dfMergeado_2020 = dfMergeado_2020.reindex(columns=['fecha', 'pm10', 'no2', 'tmed', 'prec', 'tmin', 'tmax', 'dir',
       'velmedia', 'racha', 'sol', 'presmax', 'presmin', 'intensidad'])
print(dfMergeado_2020.columns)

```

## Aplicación del modelo entrenado para obtener predicciones

```python
#Ahora preparamos el array y cargamos el Modelo entrenado

array = dfMergeado_2020.values
X = array[:,2:]
y = array[:,1]

fichero_modeloJoblib = '../AnalisisConjunto_Modelados/Madrid/Modelo_finalizado_MadridPM10.sav'
modelo_cargado = load(fichero_modeloJoblib)
resultado = modelo_cargado.score(X, y)
print("Resultado: " + str(resultado))
```

```python
predicciones = modelo_cargado.predict(X)
print(mean_squared_error(y, predicciones))

plt.figure(figsize=(21, 8))

grafica = plt.plot(y,'--r',predicciones,'-og')
plt.legend(('Valores PM10 2020', 'Predicciones PM10'),
           loc='upper right')
plt.show;
```
