### Tratamos los datos perdidos

```python
missing_data = trafico.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print('')
```

```python
missing_data = tramo.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print('')
```

```python
#reemplazamos valores perdidos por np.nan
import numpy as np
tramo.replace("?", np.NaN, inplace = True)
tramo.dropna(subset=['Tram', 'Descripció', 'Coordenades'], axis = 0, inplace= True)
tramo.reset_index(drop = True, inplace = True)
```

```python
#averiguamos que dato que se repite con más frecuencia
tramo.describe(include=[np.object])
#el top es el que se repite más veces, por lo que hay que reemplazar todos los NaN por ese en su categoría
```

```python
tramo['Descripció'].replace(np.NaN, 'Diagonal (Entença a Numància)', inplace = True)
```

```python
tramo['Coordenades'].replace(np.NaN, '2.177763706074873,41.4052353026407,2.170011548506239,41.39941440081933,2.169951058459747,41.39930869468298', inplace = True)
```

```python
#averiguamos el promedio de los integer perdidos
average_tram = tramo['Tram'].astype(float).mean(axis = 0)
```

```python
#reemplazamos los datos perdidos con este promedio
tramo['Tram'].replace(np.NaN, average_tram, inplace = True)
```
